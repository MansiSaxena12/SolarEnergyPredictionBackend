```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./SolarPrediction.csv')
```

```python
# change data column to date
df['Date'] = pd.to_datetime(df['Data'])
```

```python
df
```

```python
df.drop('Data', axis=1, inplace=True)
```

```python
df
```

```python
# convert unix time to date time
df['Time'] = pd.to_datetime(df['UNIXTime'], unit='s')
```

```python
# convert time to datetime format in TimeSunRise and TimeSunSet
df['TimeSunRise2'] = pd.to_datetime(df['TimeSunRise'], format='%H:%M:%S')
df['TimeSunSet2'] = pd.to_datetime(df['TimeSunSet'], format='%H:%M:%S')

# only keep the time part
df['TimeSunRise'] = df['TimeSunRise2'].dt.time
```

```python
df
```

```python
df_backup = df.copy()
```

```python
df.drop("Date", axis=1, inplace=True)
```

```python
df
```

```python
df.drop(['TimeSunRise2', 'TimeSunSet2', "TimeSunRise", "TimeSunSet"], axis=1, inplace=True)
```

```python
df
```

```python
df['TimeSunRise'] = df_backup['TimeSunRise']
df['TimeSunSet'] = df_backup['TimeSunSet']
```

```python
# convert time to unix time in TimeSunRise and TimeSunSet using pd.Timestamp
# lambda x where type(x) is datetime.time
df['TimeSunRise'] = df['TimeSunRise'].apply(lambda x: str(pd.Timestamp(x.strftime('%H:%M:%S')).timestamp()))
```

```python
# here x is string
df['TimeSunSet'] = df['TimeSunSet'].apply(lambda x: str(pd.Timestamp(x).timestamp()))
# df['TimeSunSet'] = df['TimeSunSet'].apply(lambda x: pd.Timestamp(x.strftime('%H:%M:%S')).timestamp())

```

```python
# use random forest to predict the data
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split


X = df.drop('Radiation', axis=1)
y = df['Radiation']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=5)
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python

model.score(X_test, y_test)
```

```python

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
# create a df to compare the actual and predicted values with x_test in the same df
df_compare = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df_compare
```

```python
# Create a copy of df1 and assign it to df2
df1 = df.copy()
```

```python
#importing vif from statmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["Column_name"] = df1.columns
vif["VIF_value"] = [variance_inflation_factor(df1.values,i) for i in range(len(df1.columns))]

#Displaying VIF values
print(vif)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate example data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create Q-Q plot
stats.probplot(data, dist="norm", plot=plt)
plt.title('Normal Q-Q plot')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```

```python
max_vif = 5

remove_flag = True

while remove_flag:
    vif=pd.DataFrame()
    vif["Column_Name"] = df1.columns
    vif["VIF_value"] = [variance_inflation_factor(df1.values,i) for i in range(len(df1.columns))]

    max_vif_column = vif.loc[vif['VIF_value'].idxmax()]

    if max_vif_column['VIF_value'] > max_vif:
        df1=df1.drop(max_vif_column["Column_Name"], axis=1)

    else:
        remove_flag = False
```
